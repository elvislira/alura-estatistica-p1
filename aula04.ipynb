```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Aula 04 - Medidas separatrizes

```python
dados = pd.read_csv('dados/dados.csv')
```

* Quartis

```python
dados.Renda.quantile([0.25, 0.50, 0.75])
```

* Decis

```python
dados.Renda.quantile(
    [i/10 for i in range(1, 10)]
)
```

* Percentis

```python
dados.Renda.quantile(
    [i/100 for i in range(1, 100)]
)
```

## Distribuição de frequências acumulada

```python
plt.figure(figsize=(14, 6))

sns.histplot(
    dados.Idade,
    cumulative=True,
    kde=True,
    color='blue'
)

plt.title('Distribuição de Frequências Acumuladas', fontsize=18)
plt.xlabel('Anos', fontsize=14)
plt.ylabel('Acumulado', fontsize=14)

plt.show()
```

```python
dados.Idade.quantile([i/10 for i in range(1, 10)])
```

## Boxplot

```python
ax = sns.boxplot(
    x = 'Altura',
    data = dados,
    orient = 'h'
)
ax.figure.set_size_inches(12, 4)
ax.set_title('Altura', fontsize=18)
ax.set_xlabel('Metros', fontsize=14)
```

```python
ax = sns.boxplot(
    x = 'Altura',
    y = 'Sexo',
    data = dados,
    orient = 'h'
)
ax.figure.set_size_inches(12, 4)
ax.set_title('Altura', fontsize=18)
ax.set_xlabel('Metros', fontsize=14)
```

```python
ax = sns.boxplot(
    x = 'Renda',
    data = dados.query('Renda < 10000'),
    orient = 'h'
)
ax.figure.set_size_inches(12, 4)
ax.set_title('Renda', fontsize=18)
ax.set_xlabel('R$', fontsize=14)
```

```python
ax = sns.boxplot(
    x = 'Renda',
    y = 'Sexo',
    data = dados.query('Renda < 10000'),
    orient = 'h'
)
ax.figure.set_size_inches(12, 4)
ax.set_title('Renda', fontsize=18)
ax.set_xlabel('R$', fontsize=14)
```

```python
ax = sns.boxplot(
    x = 'Anos de Estudo',
    data = dados,
    orient = 'h'
)
ax.figure.set_size_inches(12, 4)
ax.set_title('Anos de Estudo', fontsize=18)
ax.set_xlabel('Anos', fontsize=14)
```
