* Importando a biblioteca pandas

```python
import pandas as pd
```

* Carregando a base de dados

```python
dados = pd.read_csv('dados/dados.csv')
dados
```

## Tipos de variáveis estatísticas

>### Qualitativas

* Ordinais: Podem ser ordenadas e hierarquizadas.

```python
set(sorted(dados['Anos de Estudo']))
```

* Nominais: Não podem ser ordenadas e hierarquizadas.

```python
dados['UF'].unique()
```

```python
dados['Sexo'].unique()
```

```python
set(sorted(dados['Cor']))
```

>### Quantitativas

* Discretas: Representam uma contagem onde os valores possíveis formam um conkjunto finito ou enumerável.

```python
print(f'Idades entre {dados.Idade.min()} e {dados.Idade.max()} anos.')
```

* Contínua: Representa idade exata, sendo representado por frações de anos.

```python
print(f'De {dados.Altura.min()} até {dados.Altura.max()} metros.')
```

* Ordinal: Representa faixas de idades.
