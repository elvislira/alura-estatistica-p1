## Relação entre medidas de tendência central

* Assimetria à direita

```python
ax = sns.displot(
    dados.query('Renda < 20000').Renda, 
    kde=True
)

ax.figure.set_size_inches(12, 6)
```

```python
moda = dados.Renda.mode()
mediana = dados.Renda.median()
media = dados.Renda.mean()

print(f'Moda: {moda[0]} - Mediana: {mediana} - Média: {media}')
```

* Simétrica

```python
ax = sns.displot(dados.Altura, kde=True)
ax.figure.set_size_inches(12, 6)
```

```python
moda = dados.Altura.mode()
mediana = dados.Altura.median()
media = dados.Altura.mean()

print(f'Moda: {moda.mean()} - Mediana: {mediana} - Média: {media}')
```

* Assimetria à esquerda

```python
ax = sns.displot(dados['Anos de Estudo'], bins=17, kde=True)
ax.figure.set_size_inches(12, 6)
```

```python
moda = dados['Anos de Estudo'].mode()
mediana = dados['Anos de Estudo'].median()
media = dados['Anos de Estudo'].mean()

print(f'Moda: {moda.mean()} - Mediana: {mediana} - Média: {media}')
```
